```python
#Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Importing IRIS dataset
from sklearn import datasets
iris = datasets.load_iris()
```

```python
#To describe the iris dataset
print(iris.DESCR)
```

```python
#To view data of iris dataset
iris
```

```python
#Lets consider sepal length and sepal width as features(X). And target as (y)

X = iris.data[:,:2]
y = iris.target
```

```python
X
```

```python
y
```

```python
lab = y
```

```python
#To represent data in form of a plot

plt.scatter(X[:,0], X[:,1],c=lab)
plt.xlabel('sepal length', fontsize=15)
plt.ylabel('sepal width', fontsize=15)
```

```python
#To import KMeans Clustering

from sklearn.cluster import KMeans
```

```python
#elbow curve / SSD

ssd=[]
range_n_clusters = [2,3,4,5,6,7,8]
for n in range_n_clusters:
    kmeans = KMeans(n_clusters = n , max_iter = 50)
    kmeans.fit(X)
    
    ssd.append(kmeans.inertia_)
    
plt.plot(ssd)
```

```python
#Defining Estimator object
km = KMeans(n_clusters=3,n_jobs=4)
```

```python
#To fit the training data to the model
km.fit(X)
```

```python
#To find the centers of the 3 clusters

centers = km.cluster_centers_
centers
```

```python
new_labels = km.labels_
new_labels
```

```python
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=lab,edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels,edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```
